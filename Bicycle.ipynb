```python
import pandas as pd
import numpy as np
```

**Data Preprocessing and Data Cleaning**

```python
df.dtypes
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows:", duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna(axis=1, how='all')
```

```python
df.shape
```

```python
df.head(5)
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df.drop(["transaction_id", "product_id", "customer_id"], axis=1)
```

```python
df.describe()
```

**Data Visualization**

```python
import seaborn as sns 
import warnings
import matplotlib.pyplot as plt
%matplotlib inline

warnings.filterwarnings('ignore')
sns.set_style('darkgrid')
pd.set_option('display.max_columns', None)
```

```python
sns.distplot(df['list_price'])
```

Analisis : Berdasarkan hasil dapat dilihat ada beragam harga sepeda dengan jumlah pada kategori harga juga beragam. Dilihat dari density maka pembeli sepeda dengan harga pada rentang 1000 -1500 lebih banyak.

```python
#Brand Analysis
df['brand'].value_counts().plot(kind='bar')
```

```python
df['brand'].value_counts()
```

Sepeda dengan brand Solex adalah jumlah yang terbanyak kemudian WeareA2B dan Giant Bicycles adalah sepeda dengan jumlah kedua terbanyak, sedangkan yang lainnya seperti OHM Cycles, Trek Bicycles, dan Norco Bicycles lebih sedikit.

```python
#Average price value of each brands
sns.barplot(x=df['brand'],y=df['list_price'])
plt.xticks(rotation='vertical')
plt.show()
```

```python
#Online order analysis
df['online_order'].value_counts().plot(kind='bar')
```

```python
df['online_order'].value_counts()
```

Analisis : Transaksi secara online lebih banyak dibanding transaksi offline dengan selisih yang sangat tipis.

```python
#Average values of each type of transaction
sns.barplot(x=df['online_order'],y=df['list_price'])
plt.xticks(rotation='vertical')
plt.show()
```

Analisis : Rata-rata harga sepeda yang dibeli baik online maupun offline yaitu 1000+

```python
#Product line analysis
df['product_line'].value_counts().plot(kind='bar')
```

```python
df['product_line'].value_counts()
```

Analisis : Dilihat dari product line maka transaksi yang paling banyak yaitu pembelian sepeda standard, kedua terbanyak yaitu sepeda road, ketiga sepeda touring, dan keempat sepeda mountain

```python
#Average values of each type of product line
sns.barplot(x=df['product_line'],y=df['list_price'])
plt.xticks(rotation='vertical')
plt.show()
```

Harga rata-rata termahal yaitu sepeda touring, kedua termahal yaitu sepeda standard, ketiga yaitu sepeda road, dan yang terakhir yaitu sepeda mountain.

```python
sns.scatterplot(x=df['product_line'],y=df['list_price'])
```
